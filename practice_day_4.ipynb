## Control Structures (Conditionals and Loops)

### Level 1: Easy

#### 1. Write a program to print all the numbers from 1 to 20 using a for loop. 

```python
for x in range(1, 21):
    print(x)
```

#### 2. Create a function that prints "Hello" 10 times using a while loop. 

```python
greetings = input("Say your greetings: ")

index = 0  #initialize
while index < 10:    #check conditions
    print(greetings)
    index = index + 1  #increment
```

#### 3. Use an if-else statement to check if a given number is greater than 10. 

```python
num = 15

if num > 10:
    print("The number is greater than 10")
else:
    print("The number is less than 10")
```

#### 4. Create a program that prints the numbers 1 to 10 but skips 5 using a continue statement

```python
for i in range (1, 11):
    if i==5:
        continue
    print(i)
```

#### 5. Write a function to find the sum of all integers from 1 to a user-provided number.

```python
num = int(input("Enter the number: "))
sum = 0
for i in range(1, num+1):
    sum = sum + i
    i = i + 1
print("Sum =", sum)
```
